## Partie I. Régression linéaire simple

Calculez la matrice de corrélations.

* I.1) Si vous deviez choisir une seule variable pour réaliser une régression linéaire simple, laquelle choisiriez-vous ? Pourquoi ? 

```python
#Ecrivez votre code ici
import pandas as pd
import seaborn as sns

corrs = df.corr()
fig, ax = plt.subplots()
sns.heatmap(corrs, ax=ax, annot=True, cmap='YlGnBu', linewidths=0.2, fmt='.2f')
ax.set_title('Correlation matrix for California Housing dataset')
```

Nous voudrons créer un modèle $ \hat{y} = ax + b $, où $ \hat{y} $ est la prédiction du prix médian des maisons par quartier et $x$ est la caractéristique choisie en I.1.

* I.2) Quelles sont les valeurs optimales de $ a $ et $ b $ ? 

```python
#Ecrivez votre code ici
from sklearn.linear_model import LinearRegression
import numpy as np

X = df["MedInc"].to_numpy().reshape(-1,1)
Y = df["MedHouseVal"]

model = LinearRegression().fit(X,Y)

print(f"The coefficients(a) is: {model.coef_}")
print(f"The intercept(b) is: {model.intercept_}")
```

* I.3) Quelle est la valeur de $R^2$ ? Est-ce un bon ajustement ? Commentez \

***Réponse :*** 
La valeur de $R^2$ est loin d'être 1. Donc, il n'est pas très bon entre la valeur prédite et la valeur réelle.

* I.4) Tracez la courbe de régression sur le nuage de points entre la sortie désirée et l'entrée

```python
#Ecrivez votre code ici
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression().fit(X, Y)
Y_prediction = model.predict(X)

r_squared = r2_score(Y, Y_prediction)
print(f'R-squared = {r_squared}')

plt.scatter(X, Y, color="black", s=10, label='Observations')
plt.plot(X, Y_prediction, color="blue", linewidth=3, label= '$\hat{y} = ax+b$')

plt.xticks(())
plt.yticks(())
plt.legend()

plt.show()
```

* I.5) Quelle est l'erreur RMS de prédiction ? (en moyenne sur toutes les observations)

```python
#Ecrivez votre code ici
from sklearn.metrics import mean_squared_error
import numpy as np

RMS= mean_squared_error(Y, Y_prediction, squared=False)
print(f'RMSE = {RMS}')
```

## Partie II. Régression multilinéaire

Maintenant réalisez une régression linéaire en utilisant toutes les caractéristiques disponibles. 

* II.1) Quelle est l'erreur de prédiction ?
* II.2) Quelle est la valeur de $R^2$ ? L'ajustement est-il meilleur qu'avec une seule variable d'entrée ? \ 

***Réponse:*** La valeur de $R^2$ est proche de 1, c'est-à-dire qu'il est meilleur que la seule variable.
* II.3) Quels sont les coefficients associés aux variables d'entrée, ainsi que le terme indépendant ?

```python
#Ecrivez votre code ici
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import numpy as np

X_multi = df.drop(['MedHouseVal'], axis=1)

model_multi = LinearRegression().fit(X_multi, Y)
Y_prediction = model_multi.predict(X_multi)

RMS= mean_squared_error(Y, Y_prediction)
print("RMSE = ", np.sqrt(RMS))
r_squared = r2_score(Y, Y_prediction)
print("R_squared = ", r_squared)


for i in range(len(model_multi.coef_)):
    print(f'{model_multi.feature_names_in_[i]} = {model_multi.coef_[i]}, ')
            
print(f'Intercept = {model_multi.intercept_}')
```

Maintenant nous allons introduire un terme de régularisation, donc la nouvelle fonction de coût est : $$ J' = J + \alpha\sum_{i}{w_i^2} $$
* II.4) Quelle est la valeur optimale du paramètre $\alpha$ ? (Trouvez-la empiriquement) 
* II.5) Quelle est l'erreur RMS avec $\alpha$ optimal ?

```python
#Ecrivez votre code ici
#Using the linear CV model
from sklearn.linear_model import RidgeCV

ridge_cv = RidgeCV(alphas = [0.0001, 0.001,0.01, 0.1, 1, 10]).fit(X_multi, Y)
print(f'optimal_alpha = {ridge_cv.alpha_}')

Y_pred = ridge_cv.predict(X_multi)

RMSE= mean_squared_error(Y, Y_pred, squared=False)

print(f'RMES_ridge = {RMSE}')
```

Ajoutez des variables polynomiales d'ordre $o=2$ aux variables déjà disponibles. Comparez la régression avec et sans régularisation (RMSE, $R^2$)
* II.6) Laquelle donne des meilleurs résultats ? Pourquoi ?  

```python
#Ecrivez votre code ici
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)

X_poly = poly.fit_transform(X_multi)

model_multi = LinearRegression().fit(X_poly, Y)
Y_pred_poly = model_multi.predict(X_poly)
print("Régression sans régularisation de variables polynomiales d'ordre 2")
RMS= mean_squared_error(Y, Y_pred_poly)
print("RMSE = ", np.sqrt(RMS))
r_squared = r2_score(Y, Y_pred_poly)
print("R_squared = ", r_squared)
print("----"*20)

ridge_cv = RidgeCV(alphas = [0.0001, 0.001,0.01, 0.1, 1, 10]).fit(X_poly, Y)
print(f'optimal_alpha = {ridge_cv.alpha_}')

Y_pred_poly = ridge_cv.predict(X_poly)

RMSE= mean_squared_error(Y, Y_pred_poly, squared=False)
print("Régression avec régularisation Ridge de variables polynomiales d'ordre 2")
print(f'RMES_ridge = {RMSE}')
r_squared = r2_score(Y, Y_pred_poly)
print("R_squared = ", r_squared)
```

## Partie III. Régression non linéaire avec réseaux de neurones

Construisez et entraînez un réseau de neurones type Perceptron Multicouche (PMC) à $N_{layer} = 3$ couches et avec $m=5$ neurones dans la couche cachée.

N'oubliez pas de normaliser les données en prétraitement. 

```python
#Ecrivez votre code ici
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn import set_config

input_preprocessor = StandardScaler()
target_preprocessor = StandardScaler()

X_scaled = input_preprocessor.fit_transform(X_multi)
Y_scaled = target_preprocessor.fit_transform(Y.values.reshape(-1, 1))

m = 5
mlpmodelreg = MLPRegressor(hidden_layer_sizes=(m, ), max_iter=1000).fit(X_scaled, Y_scaled)
set_config(display='diagram')
mlpmodelreg
```


* III.1) Quelle sont les prix prédits pour les dix premières observations ?
* III.2) Quelle est l'erreur RMS d'apprentissage ?

```python
#Ecrivez votre code ici
Y_pre = mlpmodelreg.predict(X_scaled)
print(f'les prix prédits pour les dis premières observations sont {Y_pre[:10]}')
print("----"*20)
RMSE = mean_squared_error(Y_scaled, Y_pre, squared=False)
print(f'RMSE = {RMSE}')
```

Réalisez une validation croisée type *K-Fold* avec $K=10$ (n'oubliez pas de mélanger les données préalablement)

* III.3) Quelle sont les erreurs RMS d'apprentissage et de test ? (Moyenne et écart-type)

```python
#Ecrivez votre code ici
from sklearn.model_selection import cross_validate, KFold

m = 5
regmodel = MLPRegressor(hidden_layer_sizes=(m, ), max_iter=1000)

kf = KFold(n_splits=10, shuffle=True)
cv = cross_validate(regmodel, X_scaled, Y_scaled, cv=kf, return_train_score=True)

print(f"Moyenne des performance en apprentissage: {np.mean(cv['train_score'])}")
print(f"Écart-type des performance en apprentissage: {np.std(cv['train_score'])}")

print(f"Moyenne des performance en test: {np.mean(cv['test_score'])}")
print(f"Écart-type des performance en test: {np.std(cv['test_score'])}")
```

Optimisez le nombre de neurones $m$ dans la couche cachée (Piste : réalisez la validation croisée pour différentes valeurs de $m$)

* III.4) Quelle est la valeur optimale de $m$ ? Pourquoi ? 

```python
#Ecrivez votre code ici

import matplotlib.pyplot as plt
acc_train = []
acc_test = []
m_list = list(range(1,11))
kf = KFold(n_splits=10, shuffle=True)

for i in m_list:
    mlpmodel = MLPRegressor(hidden_layer_sizes=(i, ), max_iter=1000)
    cv = cross_validate(mlpmodel, X_scaled, Y_scaled, cv=kf, return_train_score=True)
    acc_train.append(np.mean(cv['train_score']))
    acc_test.append( np.mean(cv['test_score']))

plt.plot(m_list,acc_train)
plt.plot(m_list,acc_test)
plt.legend(['Training Accuracy','Testing Accuracy'])
plt.title('Moyenne des performances')
plt.show()

print("Avec les performances de test:")
m_opt = m_list[acc_test.index(max(acc_test))]
print('- m_opt =', m_opt)
print('- La moyenne des performances de la validation croisée de m_opt =', max(acc_test))
```

```python
from sklearn.model_selection import KFold, cross_validate
from sklearn.linear_model import Ridge

kf = KFold(n_splits=10, shuffle=True)

reg = LinearRegression() 

mlp_model = MLPRegressor(hidden_layer_sizes=(m,), max_iter=1000)

cv_reg = cross_validate(reg, X, Y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

cv_reg_multiple = cross_validate(reg, X_multi, Y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

cv_reg_ridge = cross_validate(Ridge(), X_multi, Y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

cv_poly = cross_validate(reg, X_poly, Y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

cv_poly_ridge = cross_validate(Ridge(), X_poly, Y, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

cv_mlp_model = cross_validate(mlp_model, X_scaled, Y_scaled, cv=kf, scoring='neg_root_mean_squared_error', return_train_score=True)

rmse_cv_reg = list(map(lambda x: -x, cv_reg['test_score']))
rmse_cv_reg_multiple = list(map(lambda x: -x, cv_reg_multiple['test_score']))
rmse_cv_reg_ridge = list(map(lambda x: -x, cv_reg_ridge['test_score']))
rmse_cv_poly = list(map(lambda x: -x, cv_poly['test_score']))
rmse_cv_poly_ridge = list(map(lambda x: -x, cv_poly_ridge['test_score']))
rmse_cv_mlp_model = list(map(lambda x: -x, cv_mlp_model['test_score']))

print('RMSE of simple linear:', rmse_cv_reg)
print('RMSE of multilinear sans régul:', rmse_cv_reg_multiple)
print('RMSE of multilinear avec régul:', rmse_cv_reg_ridge)
print('RMSE of polynomiale sans régul:', rmse_cv_poly)
print('RMSE of polynomiale avec régul:', rmse_cv_poly_ridge)
print('RMSE of non linear neural network', rmse_cv_mlp_model)
```
